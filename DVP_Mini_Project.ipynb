```python
#demo feature encoding also coverting categorical data into numbers
import pandas as pd
info={
    'Gender':['male','female','female','male','female','female'],
    'Position':['CEO','Cleaner','Employee','Cleaner','CEO','Cleaner']
}
df=pd.DataFrame(info)
print(df)
```

```python
#A LabelEncoder coverts a categorical data into a number ranging from 0 to n-1, where n is the number of classes in the variable.
#These are represented as 0,1,2 in alphabetical order.
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
gender_encoded=le.fit_transform(df['Gender'])
print (gender_encoded)
```

```python
df['encode_gender'] = gender_encoded
print(df)
```

```python
encoded_position = le.fit_transform(df['Position'])
df['encoded_position'] = encoded_position
print (df)
```
